```python
# Exercise Level 2
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
```

```python
median_age = ages[len(ages)//2]
print(median_age)
```

```python
avg_age = sum(ages)//len(ages)
print(avg_age)
```

```python
age_range = max_age - min_age
print(age_range)
```
