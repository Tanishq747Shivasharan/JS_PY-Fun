```python
# Exercise Level 2
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
```

```python
ages.sort()
min_age = ages[0]
max_age = ages[-1]
print(f"Minimum age is: {min_age}")
print(f"Maximum age is: {max_age}")
ages.append(min_age)
ages.append(max_age)
print(ages)
```

```python
median_age = ages[len(ages)//2]
print(median_age)
```

```python
avg_age = sum(ages)//len(ages)
print(avg_age)
```

```python
age_range = max_age - min_age
print(age_range)
```

```python
print(abs(min_age - avg_age))
print(abs(max_age - avg_age))
```
