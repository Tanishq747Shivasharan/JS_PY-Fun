```python
import random

def random_user_id():
    st = 'abcdefghijklmnopqrstuvwxyz1234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    user_id = ''
    for i in range(6):
        ch = random.choice(st)
        user_id = user_id + ch

    return user_id

print(random_user_id())
```

```python
def user_id_gen_by_user(nch, nid):
    st = 'abcdefghijklmnopqrstuvwxyz1234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(nid):
        user_id = ''
        for i in range(nch):
            ch = random.choice(st)
            user_id = user_id + ch
        print(user_id)

user_id_gen_by_user(int(input()), int(input()))
```

```python
from random import randint

def rgb_color_gen():
    print(f'rgb({randint(0,255)},{randint(0,255)},{randint(0,255)})')

rgb_color_gen()
```

```python
def list_of_hexa_colors(color_type, count):
    if color_type == 'hexa':
        color_list = []
        hexaDecimal_ch = '0123456789abcdef'
        color = '#'
        for i in range(count):
            color = '#'
            for i in range(6):
                ch = random.choice(hexaDecimal_ch)
                color = color + ch
            color_list.append(color)
        return color_list

print(list_of_hexa_colors('hexa', 3))
```

```python
def list_of_rgb_colors(count):
    rgb_list = []
    for i in range(count):
        r = randint(0,255)
        g = randint(0, 255)
        b = randint(0,255)
        rgb_color = f"rgb({r}, {g}, {b})"
        rgb_list.append(rgb_color)
    return rgb_list

print(list_of_rgb_colors(3))
```

```python
def generate_colors(color_type, count):
    if color_type == 'hexa':
        color_list = []
        hexaDecimal_ch = '0123456789abcdef'
        color = '#'
        for i in range(count):
            color = '#'
            for i in range(6):
                ch = random.choice(hexaDecimal_ch)
                color = color + ch
            color_list.append(color)
        return color_list
    elif color_type == 'rgb':
        rgb_list = []
        for i in range(count):
            r = randint(0,255)
            g = randint(0, 255)
            b = randint(0,255)
            rgb_color = f"rgb({r}, {g}, {b})"
            rgb_list.append(rgb_color)
        return rgb_list
    else:
        return 'Please enter valid parameters.'
        
generate_colors('hexa', 3)
generate_colors('rgb', 3)
```

```python
def shuffle_list(lst):
    random.shuffle(lst)
    return lst

l = [1,2,3,4,5]
shuffle_list(l)
```

```python
def unique_list():
    numbers = []
    for i in range(7):
        n = randint(0,9)
        if n not in numbers:
            numbers.append(n)
    return numbers

unique_list()
```
