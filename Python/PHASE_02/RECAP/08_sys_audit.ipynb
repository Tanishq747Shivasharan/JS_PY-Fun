```python
import random

def random_user_id():
    st = 'abcdefghijklmnopqrstuvwxyz1234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    user_id = ''
    for i in range(6):
        ch = random.choice(st)
        user_id = user_id + ch

    return user_id

print(random_user_id())
```

```python
from random import randint

def rgb_color_gen():
    print(f'rgb({randint(0,255)},{randint(0,255)},{randint(0,255)})')

rgb_color_gen()
```
