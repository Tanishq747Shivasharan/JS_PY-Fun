Conditionals and Booleans: Access Control Logic

```python
user_status = 'Admin'
```

```python
login_time = int(input("Enter login time: "))
```

```python
if user_status == 'Admin' and 9 < login_time < 17:
    print("Access Granted")
else:
    print("Access Denied")
```

```python
age = int(input("Enter your age: "))
if age>=18:
    print("You are old enough to learn to drive.")
else:
    print(f"You need {18-age} more years to learn to drive.")
```

```python
my_age = int(input("Enter your age: "))
your_age = int(input("Enter your age: "))
if my_age>your_age:
    print(f"You'r {my_age-your_age} years older then me!")
else:
    print(f"You'r {your_age-my_age} years younger then me!")
```

```python
num1 = int(input("Enter number one: "))
num2 = int(input("Enter number two: "))
if num1>num2:
    print(f"{num1} is greater then {num2}")
elif num2>num1:
    print(f"{num2} is greater then {num1}")
else:
    print("Both are equal")
```

```python
score = int(input("Enter you'r score: "))
if score>=90 and score<=100:
    print("A")
elif score>=80 and score<90:
    print("B")
elif score>=70 and score<80:
    print("C")
elif score>=60 and score<70:
    print("D")
elif score>=0 and score<=59:
    print("F")
else:
    print("Enter a valid score!")
```

```python
month = input("Enter a month: ")
if month == "December" or month=="January" or month=="February":
    print("Winter")
elif month=="March" or month=="April" or month=="May":
    print("Spring")
elif month=="June" or month=="July" or month=="August":
    print("Summer")
elif month=="September" or month=="October" or month=="November":
    print("Autumn")
else:
    print("Enter a valid month!")
```

```python
fruits = ['banana', 'orange', 'mango', 'lemon']
fruit = input("Enter a fruit: ")

if fruit in fruits:
    print("That fruit already exist in the list")
else:
    fruits.append(fruit)
    print(fruits)
```
