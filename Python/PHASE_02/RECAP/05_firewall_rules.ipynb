Conditionals and Booleans: Access Control Logic

```python
user_status = 'Admin'
```

```python
login_time = int(input("Enter login time: "))
```

```python
if user_status == 'Admin' and 9 < login_time < 17:
    print("Access Granted")
else:
    print("Access Denied")
```

```python
age = int(input("Enter your age: "))
if age>=18:
    print("You are old enough to learn to drive.")
else:
    print(f"You need {18-age} more years to learn to drive.")
```

```python
my_age = int(input("Enter your age: "))
your_age = int(input("Enter your age: "))
if my_age>your_age:
    print(f"You'r {my_age-your_age} years older then me!")
else:
    print(f"You'r {your_age-my_age} years younger then me!")
```

```python
num1 = int(input("Enter number one: "))
num2 = int(input("Enter number two: "))
if num1>num2:
    print(f"{num1} is greater then {num2}")
elif num2>num1:
    print(f"{num2} is greater then {num1}")
else:
    print("Both are equal")
```
