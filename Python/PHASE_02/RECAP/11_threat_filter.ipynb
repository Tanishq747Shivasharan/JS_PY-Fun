List Comprehensions: Rapid Threat Filtering

```python
scores = [21, 33, 34, 69, 80, 89, 99, 58]
```

```python
Risk_lists = [item for item in scores if item > 80]
```

```python
print(Risk_lists)
```

```python
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]
[i for i in numbers if i <= 0]
```

```python
list_of_lists =[[1, 2, 3], [4, 5, 6], [7, 8, 9]]

result = [j for i in list_of_lists for j in i]
print(result)
```

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]
result = [
    [j[0].upper(), j[0][:3].upper(), j[1].upper()]
    for i in countries
    for j in i
]

print(result)
```
