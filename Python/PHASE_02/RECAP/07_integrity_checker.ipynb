Functions: Reusable Security Tools

```python
def validate_ip(ip):
    parts = ip.split(".")
    if len(parts) == 4:
        return True
    else:
        return False
```

```python
ip = input("Enter any IPv4 address: ")
print(validate_ip(ip))
```
