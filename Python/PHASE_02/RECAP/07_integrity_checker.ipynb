```python
def validate_ip(ip):
    parts = ip.split(".")
    if len(parts) == 4:
        return True
    else:
        return False
```

```python
ip = input("Enter any IPv4 address: ")
print(validate_ip(ip))
```

```python
def add_all_nums(*args):
    total = 0
    for i in args:
        if type(i)==int or type(i)==float:    # OR isinstance(i, (int, float))
            total += i 
        else:
            print("Please enter valid numbers!")
            return
    return total
            
print(add_all_nums(2, 3,"hello", 4))
```

```python
def convert_celsius_to_fahrenheit(c):
    F = (c * 9/5) + 32
    return F

print("C converted to F :",convert_celsius_to_fahrenheit(float(input("Enter degree celsius: "))))
```

```python
def check_season(month):
    if month == "December" or month=="January" or month=="February":
        return "Winter"
    elif month=="March" or month=="April" or month=="May":
        return "Spring"
    elif month=="June" or month=="July" or month=="August":
        return "Summer"
    elif month=="September" or month=="October" or month=="November":
        return "Autumn"
    else:
        return "invalid"

print("The season is: ",check_season(input("Enter a month: ")))
```

```python
def calculate_slope(x1,y1,x2,y2):
    m = 0
    m1 = y2-y1
    m2 = x2 - x1
    if m2 == 0:
        print("Cannot divide by zero please enter appropriate values!")
    else:
        m = (y2-y1)/(x2-x1)    # Slope formula
    return m

print(calculate_slope(5, 2, 5, 10))
```

```python
def reverse_list(arr):
    reversed_arr = []

    for i in range(len(arr) - 1, -1, -1):
        reversed_arr.append(arr[i])

    return reversed_arr

print(reverse_list([1, 2, 3, 4, 5]))
print(reverse_list(["A", "B", "C"])) 
```

```python
def sum_of_odds(n):
    sum = 0
    for i in range(n+1):
        if i%2!=0:
            sum = sum + i 
    return sum

print(sum_of_odds(5))
```

```python
def sum_of_evens(n):
    sum = 0
    for i in range(n+1):
        if i%2==0:
            sum = sum + i 
    return sum

print(sum_of_evens(5))
```

```python
def evens_and_odds(n):
    count_even = 0
    count_odd = 0
    for i in range(n+1):
        if i%2==0:
            count_even += 1
        else:
            count_odd += 1
    return count_even, count_odd

print(evens_and_odds(100))
```

```python
def factorial(n):
    i = 1
    fact = 1
    while i<=n:
        fact = fact * i
        i += 1
    return fact

print(factorial(4))
```

```python
def calculate_mean(lst):
    sum = 0
    for i in lst:
        sum = sum + i
    mean = sum / len(lst)
    return mean

def calculate_median(lst):
    lst.sort()
    if len(lst) % 2 != 0:
        middle_index = lst[len(lst)//2]
        return middle_index
    else:
        mid = len(lst) // 2
        mid_1 = lst[mid-1]
        mid_2 = lst[mid]
        median = (mid_1 + mid_2) / 2
        return median

def calculate_mode(lst):
    mode = lst[0]
    max_count = lst.count(mode)
    for i in lst:
        current_count = lst.count(i)
        if current_count > max_count:
            max_count = current_count
            mode = i
    return mode

def calculate_range(lst):
    lst.sort()
    min_element = min(lst)
    max_element = max(lst)
    range = min_element + max_element
    return range

def calculate_variance(lst):
    mean = calculate_mean(lst)
    total = 0
    for i in lst:
        temp = i - mean
        temp = temp**2
        total += temp
    variance = total/len(lst)
    return variance

import math

def calculate_std(lst):
    variance = calculate_variance(lst)
    std_d = math.sqrt(variance)
    return std_d

l = [4, 6, 8, 12, 15, 4, 8, 8]
print(calculate_mean(l))
print(calculate_median(l))
print(calculate_mode(l))
print(calculate_range(l))
print(calculate_variance(l))
print(calculate_std(l))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
def unique(lst):
    for i in lst:
        if lst.count(i) == 1:
            return True
        else:
            return False

lst = [4, 5, 6, 3]
print(unique(lst))
```

```python
def type_check(lst):
    type0 = type(lst[0])
    for i in lst:
        if type0 != type(i):
            return False
    return True

lst = [4, 5, 6, 3]
l = [1, "A", 3]
print(type_check(lst))
print(type_check(l))
```
