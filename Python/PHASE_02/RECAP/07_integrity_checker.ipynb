Functions: Reusable Security Tools

```python
def validate_ip(ip):
    parts = ip.split(".")
    if len(parts) == 4:
        return True
    else:
        return False
```

```python
ip = input("Enter any IPv4 address: ")
print(validate_ip(ip))
```

```python
def add_two_numbers(firstnum, secondnum):
    sum = firstnum + secondnum
    return sum
print("The sum of two numbers is: ",add_two_numbers(5, 7))
```

```python
def area_of_circle(r):
    area = 3.14 * r * r
    return area

print("The area of circle is: ",area_of_circle(float(input("Enter radius of circle: "))))
```

```python
def add_all_nums(*args):
    total = 0
    for i in args:
        if type(i)==int or type(i)==float:    # OR isinstance(i, (int, float))
            total += i 
        else:
            print("Please enter valid numbers!")
            return
    return total
            
print(add_all_nums(2, 3,"hello", 4))
```

```python
def convert_celsius_to_fahrenheit(c):
    F = (c * 9/5) + 32
    return F

print("C converted to F :",convert_celsius_to_fahrenheit(float(input("Enter degree celsius: "))))
```

```python
def check_season(month):
    if month == "December" or month=="January" or month=="February":
        return "Winter"
    elif month=="March" or month=="April" or month=="May":
        return "Spring"
    elif month=="June" or month=="July" or month=="August":
        return "Summer"
    elif month=="September" or month=="October" or month=="November":
        return "Autumn"
    else:
        return "invalid"

print("The season is: ",check_season(input("Enter a month: ")))
```

```python
def calculate_slope(x1,y1,x2,y2):
    m = 0
    m1 = y2-y1
    m2 = x2 - x1
    if m2 == 0:
        print("Cannot divide by zero please enter appropriate values!")
    else:
        m = (y2-y1)/(x2-x1)    # Slope formula
    return m

print(calculate_slope(5, 2, 5, 10))
```
