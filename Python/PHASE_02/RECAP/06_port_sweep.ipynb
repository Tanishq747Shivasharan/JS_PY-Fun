Loops and Iterations: Automated Port Sweeping

```python
ports = [80, 24, 120, 442, 200, 22, 69, 202, 404]
```

```python
for i in ports:
    if i == 22:
        print("Critical Service Found")
        break
```

```python
# Iterate 0 to 10 using for loop, do the same using while loop.
for i in range(11):
    print(i)
i = 0
while i < 11:
    print(i)
    i = i + 1
```

```python
# for num in range(10,0, -1):
#     print(num)
num = 10
while num >= 0:
    print(num)
    num = num - 1
```

```python
# Basic syntax for patterns is !
for row in range(7): # --> Here 7 is number of rows
    for col in range(7): # --> Here 7 is number of columns
        print("#", end=" ")
    print()
```

```python
for i in range(1,8):
    for j in range(i):
        print("#", end=" ")
    print()
```

```python
for i in range(11):
    print(f"{i} x {i} = {i*i}")
```

```python
for e in range(101):
    if e%2==0:
        print(e)
```

```python
for e in range(101):
    if e%2!=0:
        print(e)
```

```python
sum = 0
for i in range(101):
    sum = sum + i
print(f"The sum of all the numbers is {sum}.")
```

```python
sum_even = 0
sum_odd = 0
for i in range(101):
    if i%2==0:
        sum_even = sum_even + i
    else:
        sum_odd = sum_odd + i
print(f"The sum of all evens is {sum_even}. And the sum of all odds is {sum_odd}.")
```
